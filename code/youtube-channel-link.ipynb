```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
from tqdm import notebook

# 윈도우 버전
# chromedriver = "C:/Users/201824893/Desktop/코딩/Python/잔재미코딩_셀레니움스크래피 강의/chromedriver.exe"

# 맥버전
chromedriver = "/Users/ojoohk/Desktop/python/chromedriver"


def driver_open(chromedriver):
    options = webdriver.ChromeOptions()
    options.add_argument("window-size=1920x1080")
    options.add_argument("headless")
    options.add_argument("lang=ko_KR")
    driver = webdriver.Chrome(chromedriver, options=options)
    return driver
```

```python
def ytb_link_scrapper(link):
    video_url = []
    driver = driver_open(chromedriver)
    driver.maximize_window()
    driver.get(link)

    time.sleep(1)
    endkey = 3 # 90~120개 / 늘릴때마다 30개
    while endkey:
        driver.find_element(By.TAG_NAME,"body").send_keys(Keys.END)
        time.sleep(1)
        endkey -= 1

    channel_name = driver.find_element(By.XPATH,'//*[@id="text"]').text
    subscribe = driver.find_element(By.XPATH,'//*[@id="subscriber-count"]').text[4:]
    print(f"{channel_name}-{subscribe}")
#     channel = f"{channel_name}-{subscribe}"

    html = driver.page_source
    soup = BeautifulSoup(html,"html.parser")

    video_grid = soup.find("ytd-rich-grid-renderer")
    video_list = soup.select('#video-title-link')

    base_url = 'http://www.youtube.com'


    for i in range(len(video_list)-1):
        url = base_url + video_list[i]["href"]
#             print(url)
        video_url.append(url)

    driver.quit()
        
    return video_url,channel_name,subscribe
```

```python
def ytb_info(video_url,channel_name,subscribe):
    print(f'{channel_name} 스크래핑 시작')
    driver = driver_open(chromedriver)
    
    date_list = []
    title_list = []
    view_list = []
    like_list = []
    comment_list = []
    url_list = []
    des_list = []
    
    for i in notebook.tqdm(range(len(video_url))):        
#     for i in notebook.tqdm(range(5)):        

        start_url = video_url[i]
        print(start_url)
        driver.get(start_url)
        time.sleep(2)
        
        driver.find_element(By.ID,"expand").click()
        
        body = driver.find_element(By.TAG_NAME,"body")
        
        num_of_pagedowns = 5
        while num_of_pagedowns:
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.5)
            num_of_pagedowns -= 1
            time.sleep(0.5)
        
        time.sleep(2)
            
        info = driver.find_element(By.ID,"above-the-fold")
        title = info.find_element(By.ID,"title").text
        bottom = info.find_element(By.ID,"bottom-row")
        view = bottom.find_element(By.CSS_SELECTOR, '#info > span:nth-child(1)').text
        if "조회수" in view:
            view = view.split()[1]
        date = bottom.find_element(By.CSS_SELECTOR, '#info > span:nth-child(3)').text
        like = info.find_element(By.XPATH, '//*[@id="segmented-like-button"]/ytd-toggle-button-renderer/yt-button-shape/button/div[2]/span').text
        comment = driver.find_element(By.CSS_SELECTOR,'#count > yt-formatted-string').text
        if "댓글" in comment:
            comment = comment.split()[1]
        description = driver.find_element(By.CSS_SELECTOR,"#description-inline-expander").text
        
        print(f"{date}/{view}/{like}/{comment}/{title}")
        
        date_list.append(date)
        title_list.append(title)
        view_list.append(view)
        like_list.append(like)
        comment_list.append(comment)
        url_list.append(start_url)
        des_list.append(description)
        
    ytb_dict = {"날짜":date_list,
                "조회수":view_list,
                "좋아요":like_list,
                "댓글 수":comment_list,
                "제목":title_list,
                "주소":url_list,
                "설명":des_list
                }
    df = pd.DataFrame(ytb_dict)
    df.to_csv(f"./{channel_name}-{subscribe}.csv")
    return df
```

```python
ytb = [
       "https://www.youtube.com/channel/UCoW9qWiUvytH7ZpnYcX4dEQ/videos"
      ]


for i in ytb:
    video_url,channel_name,subscribe = ytb_link_scrapper(i)
    ytb_info(video_url,channel_name,subscribe)
```
